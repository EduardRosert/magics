# How to plot a set up a cartesian projection with Magics

### First import Magics


```python
import Magics.macro as magics
```

###### First we have to define a cartesian projection.
This can be done use the magics.mmap object. <br/>
The documentation can be found in the  <a href='https://software.ecmwf.int/wiki/display/MAGP/Subpage+-+Projection'> Manual</a>

```python
map = magics.mmap(subpage_map_projection = "cartesian",
                  subpage_x_axis_type = 'regular',
                  subpage_y_axis_type = 'regular',
                  subpage_x_min = -100.,
                  subpage_x_max = +200.,
                  subpage_y_min = -20.,
                  subpage_y_max = 20. )
horizontal_axis = magics.maxis(axis_orientation = "horizontal",)
vertical_axis = magics.maxis(axis_orientation = "vertical",)
title = magics.mtext(text_lines=['Cartesian projection'])
magics.plot(map, horizontal_axis, vertical_axis, title)
```

###### Now we can explore the attributes of the axis : colour, grid, minor ticks etc .. 
https://software.ecmwf.int/wiki/display/MAGP/Axis

```python
horizontal_axis = magics.maxis(axis_orientation = "horizontal",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_minor_tick ='on',
                               axis_minor_grid ='on',
                               axis_minor_grid_line_style = "dot",
                               axis_minor_grid_colour = "grey",
                               )
vertical_axis = magics.maxis(axis_orientation = "vertical",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",)
title = magics.mtext(text_lines=['Cartesian projection', 'Discovering the axis attributes'])
magics.plot(map, horizontal_axis, vertical_axis, title)
```

###### You can add a title to this axis 

```python
horizontal_axis = magics.maxis(axis_orientation = "horizontal",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_minor_tick ='on',
                               axis_minor_grid ='on',
                               axis_minor_grid_line_style = "dot",
                               axis_minor_grid_colour = "grey",
                               axis_title = 'on',
                               axis_title_text = "Example of <font colour='evergreen'> coloured title </font>"
                               )
vertical_axis = magics.maxis(axis_orientation = "vertical",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_tip_title = 'on',
                               axis_tip_title_text = "Units",
                               axis_tip_title_colour = "rgb(0.16,0.38,0.51)")
magics.plot(map, horizontal_axis, vertical_axis, title)
```

###### You can define a specific list of ticks :
axis_type = 'position_list'
axis_tick_position_list --> the list of ticks you want in the system coordinates.

###### You can also use you own labels 

```python
horizontal_axis = magics.maxis(axis_orientation = "horizontal",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_minor_tick ='on',
                               axis_minor_grid ='on',
                               axis_minor_grid_line_style = "dot",
                               axis_minor_grid_colour = "grey",
                               axis_title = 'on',
                               axis_title_text = "Playing with a specific tick list...",
                               axis_type = 'position_list',
                               axis_tick_position_list = [-100., -75., 0, 100, 150., 200.],
                               axis_tick_label_type = "label_list",
                               axis_tick_label_list = ["label 1", 
                                                       "label 2", 
                                                       "<font colour='red'> origin </font>", 
                                                       "label 3", 
                                                       "label 4", 
                                                       "Final label",],
                              )
                              
vertical_axis = magics.maxis(axis_orientation = "vertical",
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_tip_title = 'on',
                               axis_tip_title_text = "Units",
                               axis_tip_title_colour = "rgb(0.16,0.38,0.51)")
magics.plot(map, horizontal_axis, vertical_axis, title)
```

######  Now let's work with some dates ..
subpage_x_axis_type = <b> date </b>

The format expected for <i>subpage_x_date_min</i> and <i>subpage_x_date_max</i> is YYYY-MM-DD HH:MM

```python
import datetime
now = datetime.datetime.today()
start_date = now.strftime("%Y-%m-%d 00:00")
later = now + datetime.timedelta(days=10)
end_date = later.strftime("%Y-%m-%d 00:00")
map = magics.mmap(subpage_map_projection = "cartesian",
                  subpage_x_axis_type = 'date',
                  subpage_y_axis_type = 'regular',
                  subpage_x_date_min = start_date,
                  subpage_x_date_max = end_date,
                  subpage_y_min = -20.,
                  subpage_y_max = 20. )
```

###### But we have to redefine the horizontal axis too .. 

```python
horizontal_axis = magics.maxis(axis_orientation = "horizontal",
                               axis_type = 'date',
                               axis_grid = "on",
                               axis_grid_line_style = "solid",
                               axis_grid_thickness = 1,
                               axis_grid_colour = "grey",
                               axis_minor_tick ='on',
                               axis_minor_grid ='on',
                               axis_minor_grid_line_style = "dot",
                               axis_minor_grid_colour = "grey",
                               axis_title = 'on',
                               axis_title_text = "Playing with time axis...",
                               )
magics.plot(map, horizontal_axis, vertical_axis, title)
```
